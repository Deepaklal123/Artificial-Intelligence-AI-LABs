```python
print('Write a Python program to find the average CGPA of your class in a dictionary.\n')
dictionary = {'Sagheer':3.8,'Hammad':3.5,'Deepak':3.1}
values=0
for i in dictionary.values():
      values += i
length = len(dictionary)
avg = values/length
print("Average CGPA    =  ",avg)
```
