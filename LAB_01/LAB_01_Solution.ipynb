```python
print(""" Q:01   Define “names” and “height” lists that contain name and height of your friends respectively. Find
the total number of entries in list, maximum height, and minimum height with their names and
find the average height of your friends """)

names = ['sagheer','hammad','saeed','sarwar','fatima','deepak']
heights = [6, 5, 6.6, 7, 5.3, 5.5]

print("Entries in names   list : ", len(names))
print("Entries in heights list : ", len(heights))

max_1 = max(heights)
min_1 = min(heights)
avg = sum(heights)/len(heights)
print('The maximum Height is of',names[heights.index(max_1)],' is ', max_1)
print('The minimum Height is of',names[heights.index(min_1)],' is ', min_1)
sum1=sum(heights)
avg=sum1/(len(heights))
print('The average of Heights is ', avg)
```

```python
print("""Q:03   Write a function that takes input a number, which is temperature in degree Celsius and return the
temperature in kelvin and Fahrenheit. """)

print('Temperature Conversion from Celcius to kelvin and Fahrenheit.')
Tc=input('Entern Temperature in celcius ')
Tc=float(Tc)
def Temp(Tc):
    Tf=(((9/5)*Tc)+32)
    print('Temperature in Fahrenheit ', Tf)
    Tk=(Tc+273)
    print('Temperature in Kelvin', Tk)
Temp(Tc)
```
