```python
#Applying logical conditions in Pandas. 

import pandas as pd
df=pd.DataFrame({'A':[10,2,3,4],'B':[3,2,24,3],'C':[1,5,6,4],'D':[6,5,4,6]})
df>3
```

```python
#Following method is used to get the values instead of Boolean values (True and False) values 

import pandas as pd
df=pd.DataFrame({'A':[10,2,3,4],'B':[3,2,24,3],'C':[1,5,6,4],'D':[6,5,4,6]})
df[df>3]
```

```python
#Filling NaN values with Pandas....with zeros 

import pandas as pd
df=pd.DataFrame({'A':[10,2,3,4],'B':[3,2,24,3],'C':[1,5,6,4],'D':[6,5,4,6]})
df= df[df > 3]
df.fillna(0)
```

```python
#Filling with mean

import pandas as pd
df=pd.DataFrame({'A':[10,2,3,4],'B':[3,2,24,3],'C':[1,5,6,4],'D':[6,5,4,6]})
df= df[df > 3]
mean = df['A'].mean()
df.fillna(mean)
```
